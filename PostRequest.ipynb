```python
import requests
import json
import numpy as np
import base64
import IPython.display as ipd

url = "http://127.0.0.1:8000/chat/"
data = {
    "user_id": 'e2efgsgf21q',
    "speaker_id": 29,
    "user_question": "一起去游乐园吧。"
}

response = requests.post(url, json=data)

if response.status_code == 200:
    result = json.loads(response.text)
    answer = result['answer']
    audio_base64 = result['audio_base64']

    audio_bytes = base64.b64decode(audio_base64)
    audio_np = np.frombuffer(audio_bytes, dtype=np.float32)

    print(f"Assistant: {answer}")
    print(f"{result['action']}")
    print(f"{result['expression']}")
    print(f"{result['expression_weight']}")
    ipd.display(ipd.Audio(audio_np, rate=22050, normalize=False))
else:
    print("Error:", response.status_code)
```

```python
import json

def find_chara_name_english_by_id(chara_id):
    with open('id_to_name/umadata.json', 'r', encoding='utf-8') as f:
        chara_list = json.load(f)

    for chara in chara_list:
        if chara['charaId'] == chara_id:
            return chara['charaNameEnglish'].replace(" ","")
    return None
```

```python
# 举个栗子
chara_id = "chara_1001_00"
chara_id = int(chara_id.split("_")[1].strip())
chara_name_english = find_chara_name_english_by_id(chara_id)
print(chara_name_english)
```
